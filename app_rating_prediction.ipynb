```python
cols_drop=['App','Last Updated','Current Ver','Android Ver']
inp1=inp1.drop(columns=cols_drop)
```

```python
# Dummy/Indicator variable
```

```python
#Category
dummy_variable_1=pd.get_dummies(inp1['Category'],dtype=int)
dummy_variable_1.head()
```

```python
#Genres
dummy_variable_2=pd.get_dummies(inp1['Genres'],dtype=int)
dummy_variable_2.head()
```

```python
# Content Rating
dummy_variable_3=pd.get_dummies(inp1['Content Rating'],dtype=int)
dummy_variable_3.head()
```

```python
inp1=pd.concat([inp1,dummy_variable_1,dummy_variable_2,dummy_variable_3],axis=1)
inp1.columns
```

```python
inp1.drop(['Category','Genres','Content Rating'],axis=1,inplace=True)
```

```python
inp1.head()
```

```python
# Type
dummy_variable_3=pd.get_dummies(inp1,columns=['Type'],dtype=int)
dummy_variable_3.head()
```

```python
dummy_variable_3.drop('Type_Free',axis=1,inplace=True)
```

```python
dummy_variable_3.head()
```

```python
inp1=pd.concat([inp1,dummy_variable_3],axis=1)
inp1.drop('Type',axis=1,inplace=True)
inp1.head()
```

```python
# Train test split
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
# target variable=Rating (y)
y_data=inp1['Rating']
x_data=inp1.drop('Rating',axis=1)
x_data.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.3,random_state=42)
print("The elements in training dataset is:",x_train.shape[0])
print("The elements in testing dataset is:",x_test.shape[0])
```

```python
# Model Building
```

```python
lnr=LinearRegression()
lnr.fit(x_train,y_train)
```

```python
y_train_pred=lnr.predict(x_train)
y_test_pred=lnr.predict(x_test)
```

```python
train_mae=mean_absolute_error(y_train,y_train_pred)
test_mae=mean_absolute_error(y_test,y_test_pred)
print("The training mean absolute error is:",train_mae)
print("The testing mean absolute error is:",test_mae)
```

```python
r2_score_train=r2_score(y_train,y_train_pred)
r2_score_test=r2_score(y_test,y_test_pred)
print("The r2 score for training dataset is:",r2_score_train)
print("The r2 score for testing dataset is:",r2_score_test)
```
